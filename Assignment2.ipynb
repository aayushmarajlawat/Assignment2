```python
# 1. Create a list with 3 fruits (for example: apple, banana, orange).
# • Print the entire list.
# • Print each item separately.

fruits = ["apple", "banana", "orange"]
print("Entire list:", fruits)
print("Entire list separately:")
print(fruits[0])
print(fruits[1])
print(fruits[2])
print("\n","\n")

# 2. Create a list with 4 names.
# • Change the first name to a different one.
# • Print the new list.

names = ["Ram", "Sita", "Gita", "Hari"]
names[0] = "Eve"
print("New names list:", names)
print("\n","\n")

# 3. Create a list with 5 vehicles (for example: car, bus, bike).
# • Add one more vehicle to the end of the list.
# • Print the list.

vehicles = ["car", "bus", "bike", "train", "truck"]
vehicles.append("scooter")
print("Vehicles list:", vehicles)
print("\n","\n")

# 4. Create a list with 6 foods.
# • Remove one food from the list.
# • Print the new list.

foods = ["pizza", "burger", "pasta", "sushi", "salad", "biryani"]
foods.remove("pasta")
print("New food items:", foods)
print("\n","\n")

# 5. Create a list with 4 cities.
# • Sort the list in alphabetical order.
# • Reverse the list.

cities = ["London", "Kathmandu", "Tokyo", "New York"]
cities.sort()
print("Cities sorted alphabetically:", cities)
cities.reverse()
print("Cities reversed:", cities)
print("\n","\n")


# 6• Create a list with 3 animals.
#  • Use a for loop to print each animal.
animals = ["dog", "cat", "rat"]
print("Animals:")
for animal in animals:
  print(animal)
print("\n","\n")

# 7• Create a list with numbers from 1 to 5.
# • Use a for loop to print each number multiplied by 2.
numbers = [1, 2, 3, 4, 5]
print("Numbers multiplied by 2:")
for num in numbers:
    print(num * 2)
print("\n","\n")

# 8. Create a list with 6 friends.
# • Print the first three friends.
# • Print the last three friends.

friends = ["Ava", "Ben", "Harry", "Elsa", "Anna", "Gita"]
print("First three friends:", friends[:3])
print("Last three friends:", friends[-3:])
print("\n","\n")

# 9.Use range() to create a list with numbers from 1 to 10.
# • Print the list.
range_num = list(range(1, 11))
print("Numbers from 1 to 10:", range_num)
print("\n","\n")

# 10• Use list comprehension to create a list with the squares of numbers from 1 to 5.
# • Print the list.
squares = [x**2 for x in range(1, 6)]
print("Squares from 1 to 5:", squares)
print("\n","\n")
```
